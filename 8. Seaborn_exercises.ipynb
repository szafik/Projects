### TASK: Recreate the Scatter Plot shown below

**The scatterplot attempts to show the relationship between the days employed versus the age of the person (DAYS_BIRTH) for people who were not unemployed. ** 

<img src="task_one.jpg">


```python
df_employed = df[df['OCCUPATION_TYPE'].isna() == False]
```

```python
df_employed[['DAYS_BIRTH', 'DAYS_EMPLOYED']] = df_employed[['DAYS_BIRTH' , 'DAYS_EMPLOYED']].abs()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df_employed, x='DAYS_BIRTH', y='DAYS_EMPLOYED', alpha=0.01)
```

### TASK: Recreate the Categorical Plot shown below:

<img src='catplot_solution.png'>

**This plot shows information only for the *bottom half* of income earners in the data set.  Note: You will need to adjust or only take part of the dataframe *before* recreating this plot.**

```python
bottom_half_income = df.nsmallest(n=int(len(df)/2), columns="AMT_INCOME_TOTAL")
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data= bottom_half_income, x="NAME_FAMILY_STATUS", y="AMT_INCOME_TOTAL", hue="FLAG_OWN_REALTY")
plt.legend(loc=(1.05, 0.5), title="FLAG_OWN_REALTY")
```
